```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = "data_predicted.csv"
data = pd.read_csv(file_path)

# Calculating the absolute error between HR and HR_predicted
data['error'] = abs(data['HR'] - data['HR_predicted'])

# Group by 'Condition' and calculate the mean error for each condition
mean_errors = data.groupby('Condition')['error'].mean()

print("Mean error for each condition:")
print(mean_errors)

plt.figure(figsize=(10, 6))
mean_errors.plot(kind='bar', color='skyblue')
plt.title('Mean Error by Condition', fontsize=16)
plt.ylabel('Mean Error', fontsize=14)
plt.xlabel('Condition', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()
plt.show()
```
